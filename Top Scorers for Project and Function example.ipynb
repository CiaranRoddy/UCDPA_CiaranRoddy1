```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("EPL_20_21.csv")
```

```python
# slicing to view goals


df.loc[:,"Name":"Goals"]
```

```python
# getting the data for player who have 15 or more goals 


g=df[df.Goals > 15]
```

```python
g
```

```python
# displaying them in ascending order


ng=g.loc[:,["Name","Goals"]]
ngg=ng.sort_values(by="Goals",ascending=False)
```

```python
ngg
```

```python

# graph to illustrate the League's top scorers


x=ngg["Name"]
y=ngg["Goals"]


        
plt.bar(x,y, width=0.9,color = ["honeydew","red","tomato","violet","yellow","dodgerblue"],edgecolor = 'black')
plt.xlabel("Players")
plt.ylabel("Goals")
plt.title("Premier League top scorers")
plt.xticks(rotation=75)
plt.show()
```
