```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns 
```

```python
df=pd.read_csv("EPL_20_21.csv")
```

```python
df
```

```python
df1=df.drop_duplicates(subset=["Name"])
```

```python
df1
```

```python
Age=df1["Age"]
```

```python
mean_age = Age.mean()
```

```python
plt.hist(Age, bins=5,color="aqua", edgecolor='black')
plt.axvline(mean_age, color='red', label='Mean Age', linestyle='dashed',linewidth=1)
plt.legend(loc="upper left")
plt.ylabel("Players")
plt.xlabel("Age")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.1)
plt.title("Age Histogram");
```

```python
sns.distplot(df1.Age)
plt.legend(loc="upper left")
plt.ylabel("Players")
plt.xlabel("Age")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.1)
plt.title("Age Histogram");
```
