```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("EPL_20_21.csv")
```

```python
df
```

```python
# Define the variables to plot on a box graph to illuistrate the age range of certain clubs

x=df['Club']

y=df['Age']
```

```python

plt.figure(figsize=(45,15))



b = sns.boxplot(x,y)
b.set_xlabel("Club",fontsize=40)
b.set_ylabel("Age" ,fontsize=45)
plt.xticks(rotation=90, size = 25)
plt.yticks(size = 25)
plt.title('Plot of Age range for each Club',fontsize = 40)
```
