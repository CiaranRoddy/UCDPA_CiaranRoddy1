```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df=pd.read_csv("EPL_20_21.csv")
```

```python
df
```

```python
# Dictionary used for Nationalies to get the 20 most common in the EPL


nationality_dict = dict(Counter(df['Nationality']).most_common(20))
plt.bar(nationality_dict.keys(), nationality_dict.values(), color=["r","b","yellow","coral","r","g","steelblue","orange","r","r","w","skyblue","palegreen","g","w","r","r","g","b","yellow"], edgecolor= "black")

plt.xlabel('Nationality')
plt.grid(True, color = "grey", linewidth = "0.3", linestyle = "-.")
plt.ylabel('Frequency')
plt.title('Bar Plot of Twenty Most Common Nationalities')

plt.xticks(rotation=90)

plt.show()
```
