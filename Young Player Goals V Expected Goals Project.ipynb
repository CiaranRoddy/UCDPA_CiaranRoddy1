```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("EPL_20_21.csv")
```

```python
df
```

```python
# Filtering young players by under the age of 24 & adding column of actual goals Per Game with iterrows

yg=df[(df["Age"]<24)&(df["Goals"]>1)] 
for ind,row in yg.iterrows():
    yg.loc[ind,"Actual_Goals_Per_Game"] = row['Goals'] / row['Matches']
```

```python
yg.head()
```

```python
#Analsying the young players Goals Data from Top to Bottom
# 
yg.sort_values(by="Goals",ascending=False)
```

```python
# Comparing the Actual v Expected Goals Data on Scatter Graph



x=yg["xG"]
y=yg["Actual_Goals_Per_Game"]


plt.scatter(x, y, s=100,alpha=0.35 ,edgecolor ='black', linewidth =1)


plt.ylabel('Expected Goals')

plt.xlabel('Actual Goals')


cbar=plt.colorbar()
cbar.set_label

plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)

plt.show()
```
